```python
category = ['Workclass', 'Education', 'Marital_status', 'Occupation', 'Relationship', 'Race', 'Sex', 'Native_country']
for c in category:
    un=pd.unique(df[c])
    nu=list(range(len(un)))
    df[c]=df[c].replace(un,nu)
```
